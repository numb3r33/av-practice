```python
import numpy as np
import pandas as pd
```

```python
students = pd.read_csv('../data/Student.csv')
```

```python
students.head()
```

```python
def feature_engineer(df):
    df['num_experience'] = df.shape[0]
    df['num_exp_in_job'] = (df.Experience_Type  == 'job').sum()
    df['num_awards'] = (df.Experience_Type  == 'award').sum()
    df['num_previous_internships'] = (df.Experience_Type == 'internship').sum()
    
    return df

students_subset = students.head(20)
students_subset = students_subset.groupby(['Student_ID']).apply(feature_engineer)
students_subset = students_subset.groupby('Student_ID').first()
```
